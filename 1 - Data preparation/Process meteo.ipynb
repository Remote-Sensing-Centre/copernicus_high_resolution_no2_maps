```python
import pandas as pd
```

```python
DATA_FILE = "meteo_hourly_2023_2024_GZM.csv"
OUTPUT_FILE = "meteo_2023_2024_processed.csv"
COLUMNS = [
    "datetime",
    "temperature",
    "relative_humidity",
    "surface_pressure",
    "dew_point",
    "precipitation",
    "wind_speed",
    "wind_direction",
    "u_component_of_wind_10m",
    "v_component_of_wind_10m",
]
```

```python
df = pd.read_csv(DATA_FILE)
df.head()
```

```python
df = df[COLUMNS]
df.head()
```

```python
df["datetime"] = pd.to_datetime(df["datetime"])
df.head()
```

```python
df.rename(
    columns={
        "datetime": "timestamp_utc",
        "surface_pressure": "pressure",
        "u_component_of_wind_10m": "wind_u",
        "v_component_of_wind_10m": "wind_v",
    },
    inplace=True
)
df.head()
```

```python
df.to_csv(OUTPUT_FILE, index=False)
```
